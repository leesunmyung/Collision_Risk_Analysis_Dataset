## '운전자_신발 폴더' 에서 Chassis, mobileye-obstacle 데이터 로드

```python
init_folder = '0509_' + driver_shoes + '_temp'

os.chdir(init_folder)
os.getcwd()

chassis_folder = driver_shoes + '_Chassis'
mobileye_folder = driver_shoes + '_mobileye-obstacle'

os.chdir(chassis_folder)
chassis_files = os.listdir('.')
chassis_files.sort()

os.chdir('../')
os.chdir(mobileye_folder)
mobileye_files = os.listdir('.')
mobileye_files.sort()

print(driver_shoes + '폴더\n')
if (len(chassis_files) == 12) and (len(mobileye_files) == 12):
    print("Chassis 데이터")
    print(chassis_files)
    print("Mobileye 데이터")
    print(mobileye_files)
else:
    print("다시 확인하세요.")

os.chdir('..')
os.getcwd()
```

```python
# for문 기준: 인덱스(0~11)

os.chdir(path = drive_path + init_folder + '\\')
os.getcwd()


for i in range(0, 12):
    # path = drive_path + init_folder + '\\'+ Chassis_folder + '\\'
    print(chassis_files[i], mobileye_files[i])
    chassis_data = pd.read_csv(chassis_folder + '\\' + chassis_files[i])
    mobileye_data = pd.read_csv(mobileye_folder + '\\' + mobileye_files[i])
    
    # Framing: Chassis Data
    df1 = pd.DataFrame(chassis_data)
    dfc1 = df1[['.header.seq', '.steering_angle', '.accel_pedal', '.break_pressure', '.wheel_speed', '.lat_accel', '.long_accel']]
    print(dfc1)
    # Framing: Mobileye Data
    df5 = pd.DataFrame(mobileye_data)
    print(df5)
    
    # 구간별
    for j in range(0, 3): 
        
        # Chassis Data
        # 시작 시퀀스~끝 시퀀스에서 1/5만큼 가져오기
        print("Chassis 시작 seq: " + str(start_cha_seq[j][i]) + ", 마지막 seq: " + str(end_cha_seq[j][i]))
        dfr2 = dfc1[(dfc1['.header.seq'] >= int(start_cha_seq[j][i])) & (dfc1['.header.seq'] <= int(end_cha_seq[j][i])) & (dfc1['.header.seq'] % 5 == 0)]
        print(dfr2)
        
        # 필요 attribute만 가져오기
        steering_angle = dfr2['.steering_angle']
        accel_pedal = dfr2['.accel_pedal']
        break_pressure = dfr2['.break_pressure']
        wheel_speed = dfr2['.wheel_speed']
        lat_accel = dfr2['.lat_accel']
        long_accel = dfr2['.long_accel']
        
        
        
        # Mobileye Data
        # 객체 메시지 분리: 시작 row~끝 row만큼 가져오기
        try: 
            start_row = ((df5.index[df5['.header.seq'] == int(start_mob_seq[j][i])]).tolist())[0] + 2
            end_row = ((df5.index[df5['.header.seq'] == int(end_mob_seq[j][i])]).tolist())[0] + 2
        except: 
            print("다시 입력하세요.")
        
        df6 = df5[(df5.index >= start_row) & (df5.index <= end_row)]

        # 필요 attribute만 가져오기
        obstacle_pos_x = (df6[df6['time'].str.contains('pos_x')])['time']
        obstacle_rel_vel_x = (df6[df6['time'].str.contains('rel_vel_x')])['time']

        print("mobileye 시작 row: " + str(start_row) + ", 마지막 row: " + str(end_row))
        print(obstacle_pos_x, '\n')
        print(obstacle_rel_vel_x)
        
        
        # Chassis Data + Mobileye Data => (NEW) CRP Input Data
        course = '2' if j == 0 else '4' if j == 1 else '5'
        
        crp_file = '220509_' + driver_shoes + '_' + course + '_' + str(i+1).zfill(2)    # CRP 파일명 형식: 0509_EJ_DS_2_01.csv

        with open('CRP_0509_EJ_DS\\' + crp_file + '.csv', 'w', newline="") as temp_file:
            header = ["Time", "Vs", "CalVs", "Engine", "Throttle", "Turbine", "GearStatus", "GearChange", "BrakeForce", "LongAcc", "LatAcc", "YawRate1", "YawRate2", "SAS", "Range1", "RelSpeed1", "Angle1", "Status1", "Stationary1", "Range2", "RelSpeed2", "Angle2", "Status2", "Stationary2", "Range3", "RelSpeed3", "Angle3", "Status3", "Stationary3", "Range4", "RelSpeed4", "Angle4", "Status4", "Stationary4", "Range5", "RelSpeed5", "Angle5", "Status5", "Stationary5", "Range6", "RelSpeed6", "Angle6", "Status6", "Stationary6", "Range7", "RelSpeed7", "Angle7", "Status7", "Stationary7", "Range8", "RelSpeed8", "Angle8", "Status8", "Stationary8", "Range9", "RelSpeed9", "Angle9", "Status9", "Stationary9", "Range10", "RelSpeed10", "Angle10", "Status10", "Stationary10", "Range11", "RelSpeed11", "Angle11", "Status11", "Stationary11", "Range12", "RelSpeed12", "Angle12", "Status12", "Stationary12", "RadarStatus", "RadarCurv", "TrackNo", "RadarYaw", "LaDistance", "LaVrel", "LaCurv", "LaTargetON", "LaStatus", "LaStep"]
            writer = csv.DictWriter(temp_file, fieldnames=header)

            writer.writeheader()

            time = 0
            for a, b, k, l, m, n, o, p in zip(wheel_speed, accel_pedal, break_pressure, long_accel, lat_accel, steering_angle, obstacle_pos_x, obstacle_rel_vel_x):
                int_a = int(a)
                o = o.split(' ')[1]
                p = p.split(' ')[1]
                writer.writerow({"Time": time, "Vs": int_a, "CalVs": a, "Engine": 1800, "Throttle": b, "Turbine": 1800, "GearStatus": 4, "GearChange": 0, "BrakeForce": k, "LongAcc": l, "LatAcc": m, "YawRate1": 0, "YawRate2": 0, "SAS": n, "Range1": o, "RelSpeed1": p, "Angle1": 0, "Status1": 0, "Stationary1": 0, "Range2": 0, "RelSpeed2": 0, "Angle2": 0, "Status2": 0, "Stationary2": 0, "Range3": 0, "RelSpeed3": 0, "Angle3": 0, "Status3": 0, "Stationary3": 0, "Range4": 0, "RelSpeed4": 0, "Angle4": 0, "Status4": 0, "Stationary4": 0, "Range5": 0, "RelSpeed5": 0, "Angle5": 0, "Status5": 0, "Stationary5": 0, "Range6": 0, "RelSpeed6": 0, "Angle6": 0, "Status6": 0, "Stationary6": 0, "Range7": 0, "RelSpeed7": 0, "Angle7": 0, "Status7": 0, "Stationary7": 0, "Range8": 0, "RelSpeed8": 0, "Angle8": 0, "Status8": 0, "Stationary8": 0, "Range9": 0, "RelSpeed9": 0, "Angle9": 0, "Status9": 0, "Stationary9": 0, "Range10": 0, "RelSpeed10": 0, "Angle10": 0, "Status10": 0, "Stationary10": 0, "Range11": 0, "RelSpeed11": 0, "Angle11": 0, "Status11": 0, "Stationary11": 0, "Range12": 0, "RelSpeed12": 0, "Angle12": 0, "Status12": 0, "Stationary12": 0, "RadarStatus": 0, "RadarCurv": 5000, "TrackNo": 1, "RadarYaw": 1, "LaDistance": 30, "LaVrel": 0, "LaCurv": 0, "LaTargetON": 1, "LaStatus": 24, "LaStep": 0})
                time = time + 0.05
                
            print(crp_file + ' 구성 완료\n')    
            print('========================================\n')
        
        
        
```
